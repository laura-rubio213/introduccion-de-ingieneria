cree la funcion
`create_and_select_dataframe` , en la cual se cree una dataframe a partir de un diccionario que tenga los atributos ` product `
y `price`, los cuales son arreglos con valores tipo de cadena de texto y entero segun el orden.
luego, retorne una tupla con la columna `product` y otra columna con el los precios que sean superiores a 150

````
def create_and_select_dataframe():
    df = pd.DataFrame({
        "product": [ "A", "B", "C"],
        "price": [123, 234, 150]
  })
    return df["product"], df[df["price"] > 150]

products, price = create_and_select_dataframe()
print("productos: \n", products, end="\n\n")
print("productos con un valor mayor a 150: \n", price)
````

````
def test_create_and_select_dataframe():
  products, price = create_and_select_dataframe()
  assert len(products) == 3
  assert price.iloc[0]["product"] == "B"
test_create_and_select_dataframe
````

crea una funcion llamada `add_discount_column`, la cual recibe un dataframe, una tasa de descuento, y se retorna un nuevo dataframe al cual se le añande la columna `discount`

````
def add_discount_column(df, discount_rate=0.1):
  df["discounted"] = df["price"] * (1 - discount_rate)
  return df

def test_add_discount_column():
  df = pd.DataFrame({"price": [10, 20 , 30, 50]})
  df = add_discount_column(df, 0.2)
  assert df["discounted"].tolist() == [8, 16, 24, 40]

test_add_discount_column()

````

````

def create_series_transform():
    return pd.Series([1, 5, 15, 20, 35, 14, 12, 25, 9, 45]).apply(lambda x: x * 2 if x > 15 else None).dropna()
print(create_series_transform())
````

##pistas:##

2. la serie derivada se genera de filtrar la columna de calificaciones con los valores >= 60
3. para la nueva columna, se debe tener en cuenta la serie derivada y usar la funcion `map` para determinar si aprueba o reprueba segun el estado de la calificacion
4. para contar la cantidad de aprovados y reprobados usar `value.counts()`

````

data = {
    'name': ['Ana', 'Luis', 'Pedro', 'Marta'],
    'grade': [85, None, 55, 70],
    'status': [None, None, None, None]
}

students_df = pd.DataFrame(data)
processed_df, counts = process_student_data(students_df)

print("Nuevo DataFrame:")
print(processed_df)
print("\nConteo de estados:")
print(counts)
````

````
import pandas as pd

def process_student_data(df):
    df['grade'].fillna(df['grade'].mean(), inplace=True)
    df['status'] = df['grade'].apply(lambda x: 'aprobado' if x >= 60 else 'reprobado')

    return df, df['status'].value_counts()


data = {
    'name': ['Ana', 'Luis', 'Pedro', 'Marta'],
    'grade': [85, None, 55, 70]
}


students_df = pd.DataFrame(data)
processed_df, counts = process_student_data(students_df)


print("Nuevo DataFrame:")
print(processed_df)
print("\nConteo de estados:")

````
